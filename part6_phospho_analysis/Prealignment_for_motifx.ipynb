```python
#import numpy as np
import csv
import pandas as pd
```

```python
# this function takes a string and returns all uppercases
# except for originally lowercase 's' and 't'
def convertPep(s):
    position = 0
    output = ''
    for c in s:
        if c.islower():
            #print(c)
            if c == 's' or c == 't' or c == 'y':
                #print('keep location: ', position)
                output = output + c
            else:
                #print('replace location: ', position)
                output = output + c.upper()
        else:
            output = output + c
        position += 1
    #print(output)
    return output
```

```python
# this function returns position of the first phosphorylated 's' or 't' or 'y'
def locatePhos(s):
    position = 0
    for c in s:
        if c.islower():
            return position
        else:
            position += 1
            continue
```

```python
# this function is to center the lowercase 's' or 't'

def centerST(s):
    peplength = len(s)
    phoslocation = locatePhos(s)
    
    # build the first 7aa
    # if there are >=7aa before phos, copy them; else add _ in front upto 7aa
    res_to_ST = s[0:phoslocation]
    if len(res_to_ST) >= 7:
        res_to_ST = res_to_ST[-7:]
    else:
        num_spacef = 7 - len(res_to_ST)
        res_to_ST = num_spacef*'_' + res_to_ST
    #print(res_to_ST)    
        
    
    # build the last 7aa
    # if there are >=7aa after phos, copy them; else add _ after until 7aa
    res_after_ST = s[phoslocation:len(s)+1]
    if len(res_after_ST) >= 8:
        res_after_ST = res_after_ST[0:8]
    else:
        num_spaceaf = 8 - len(res_after_ST)
        res_after_ST = res_after_ST + num_spaceaf*'_'
    #print(res_after_ST)
    
    output = res_to_ST + res_after_ST
    #print(output)
    return output
```

```python
# read .txt file
# All pepMod to be convert
# header = row number to be used as column name

input_vector = pd.read_csv('/Users/vasdumrong/Box/VasD_projects/Project_STRaxon/analysis/phospho_analysis_motif/modPep_all_pythoninput.txt', 
                           delimiter = '\t', header = 0)

input_vector = input_vector.values.tolist()
input_vector = [item for sublist in input_vector for item in sublist]
input_vector
```

```python
# generate output
centered_modPep = list(map(centerST ,map(convertPep, input_vector)))
centered_modPep
```
